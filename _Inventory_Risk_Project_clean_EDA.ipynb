```python
n_shipments = 5000

shipments = pd.DataFrame({
    'Shipment_ID': [f'SH{i:05d}' for i in range(1, n_shipments+1)],
    'Supplier_ID': np.random.choice(
        suppliers['Supplier_ID'],
        n_shipments
    ),
    'Product_ID': np.random.choice(
        products['Product_ID'],
        n_shipments
    ),
    'Warehouse_ID': np.random.choice(
        warehouses['Warehouse_ID'],
        n_shipments
    ),
    'Shipment_Date': pd.to_datetime(
        np.random.choice(
            pd.date_range(
                '2025-01-01',
                '2025-12-31'
            ),
            n_shipments
        )
    ),
    'Lead_Time_Days': np.random.randint(
        2,
        15,
        n_shipments
    ),
    'Delivery_Status': np.random.choice(
        ['On Time','Delayed'],
        n_shipments,
        p=[0.8,0.2]
    )
})

shipments.head()
```
